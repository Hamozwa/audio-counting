```python
import matplotlib.pyplot as plt
import csv

def csv_counts_histogram(csv_file, max_count=None):
    counts = []
    with open(csv_file, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            count = float(row["count"])
            if max_count is None or count <= max_count:
                counts.append(count)
    
    plt.hist(counts, bins=range(int(min(counts)), int(max(counts)) + 2), edgecolor='black', align='left')
    plt.xlabel("Label")
    plt.ylabel("Number of files")
    plt.show()
```

```python
csv_counts_histogram("CountixAV_test.csv")
csv_counts_histogram("CountixAV_test.csv", max_count=8)
```
